```python
import sqlite3
import pandas as pd
```

```python
# Turn dataframe into database
```

```python
con = sqlite3.connect("employees_df")
cur = con.cursor()
```

```python
employees_df = pd.read_csv("Resources/Employee.csv")
employees_df
```

```python
cur.execute('CREATE TABLE IF NOT EXISTS Employees (Education TEXT, JoiningYear TEXT, PaymentTier INTEGER, Age INTEGER, Gender TEXT, EverBenched TEXT, ExperienceInCurrentDomain INTEGER, LeaveOrNot INTEGER)')
con.commit()
```

```python
employees_df.to_sql('Employees', con, if_exists='replace', index = False)
```

```python
cur.execute('''  
SELECT * FROM Employees
          ''')
```

```python
for row in cur.fetchall():
    print (row)
```
